The *CHEOPS analysis* notebooks model the *CHEOPS* secondary eclipse observations using the same phase curve model as the *external analysis* notebooks. The model is implemented by `src.cheopslpf.CHEOPSLPF`, which again inherits all its functionality from `pytransit.lpf.PhaseCurveLPF`, but differs from `ExternalDataLPF` in the data initialisation, priors, and how the systematics are modelled.

- *Priors:* The orbital parameters are constrained using priors created from the external analysis posteriors.

- *Systematics:* The systematics for each CHEOPS visit are modelled using a linear model with a varying number of covariates.

The *CHEOPS* analyses are done to create a set of detrended light curves for the final joint analysis and to study how sensitive our results are on the way we constrain the EV amplitude. While the TESS and Spitzer photometry cover the whole phase of KELT-1b orbit, the CHEOPS photometry covers only the near vicinity of the secondary eclipse. This means that we can't estimate the ellipsoidal variations or Doppler boosting independently for the CHEOPS passband.

Doppler boosting doesn't create any major problems, but the measured eclipse depths (and thus the companion-star flux ratios) are very sensitive on the EV signal amplitude. Thus, we need to take a good look at the different EV scenarios.

As with the external data analyses, the first *CHEOPS* analysis case is with a theoretical prior on the EV amplitude.

```python
from multiprocessing import Pool
from src import CHEOPSLPF
```

```python
lpf = CHEOPSLPF('a')
```

```python
def lnposterior(pv):
    return squeeze(lpf.lnposterior(pv))

pool = Pool(processes=16)
```

```python
lpf.ps
```

## Find the global posterior mode

```python
lpf.optimize_global(150, npop=400, pool=pool, vectorize=False, lnpost=lnposterior)
```

```python
lpf.optimize_global(15000, npop=400, pool=pool, vectorize=False, lnpost=lnposterior)
```

```python
lpf.plot_folded_light_curve();
```

## Sample the posterior using MCMC

```python
lpf.sample_mcmc(5000, 100, repeats=10, save=True, pool=pool, vectorize=False, lnpost=lnposterior)
```

```python
df = lpf.posterior_samples()
```

```python
df.log10_ted_cheops.hist(bins=50)
```

```python
(10**df.log10_ted_cheops).hist(bins=50)
```
